```python

#  Firstly, let's simply create world map
world_map= folium.Map()

world_map
```

```python
# 37.09
# Let's define the world map centered around USA with the zoom level.
world_map = folium.Map(location=[37.09, -95.71], zoom_start=4)

world_map
```

```python
# Different map style of world map. I'm using 'Stamen Toner' for highlighting river meanders and costal zones beacause this shows high-contrast of B+W (black and white) maps.

world_map= folium.Map(location= [37.09, -95.71], zoom_start= 4, tiles= 'Stamen Toner')

world_map

```

```python
# Next is 'Stamen Terrain Maps' this feature shows hill shading and natural vegetation colors.

world_map = folium.Map(location=[37.09, -95.71], zoom_start=4, tiles='Stamen Terrain')

# display map
world_map
```
