Q2. Finding the number of times when the 'weather is exactly clear':

```python
# Three ways to solve this problem by; 
# 1. .value_counts()
# 2. filtering
# 3. groupby()

df['Weather'].value_counts()

# We can see the result: There is 1326 results are with clear weather.
```

```python
# Filtering
df[df.Weather == 'Clear']

# We can see the all 1326 records with clear weather only.
```

```python
# Groupby()
df.groupby('Weather').get_group('Clear')
```
