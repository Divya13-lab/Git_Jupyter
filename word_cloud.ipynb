```python
file_content=open ("Textwordcloud.txt").read()
```

```python
wordcloud = WordCloud(background_color='white',
                      width=1200,
                      height=1000)


# generate the word cloud
wordcloud.generate(file_content)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Word cloud with the images

file_content=open ("Textwordcloud.txt").read()
python_mask=np.array(PIL.Image.open("wordcloudmask.png"))

wordcloud = WordCloud(background_color='white',
                      width=120,
                      height=100,
                      mask=python_mask,
                      contour_color='black',
                      contour_width=3)


# generate the word cloud
wordcloud.generate(file_content)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
